Relax and Enjoy Lake Development Corporation is developing a lakeside community at a privately owned lake. They need to use various media to reach out to the respective audience. The following are the alternatives of the five media that they have chosen. Please refer to the table above:

the Management has provided an advertising budget of 30000 d for the first month campaign

At least 10 television commercial must be used

At least 50000 potential customers must be reached

No more than 18000 may be spent on tv commercials.

What advertising media selection plan should be recommended. ( Here the aim is to maximize exposure quality which takes into account demographics and psychographics of the customers)

```python
import pulp
```

```python
# Create a problem variable
prob = pulp.LpProblem('Media_Mix',pulp.LpMaximize)
#prob = pulp.LpProblem('Electronic_Communication',pulp.LpMaximize)
```

```python
# Decision variables
x = pulp.LpVariable("DTV",lowBound=0,cat='Integer')
y = pulp.LpVariable('ETV',lowBound=0,cat='Integer')
z = pulp.LpVariable('DN',lowBound=0,cat='Integer')
v = pulp.LpVariable('SN',lowBound=0,cat='Integer')
w = pulp.LpVariable('R',lowBound=0,cat='Integer')
variables = []
# should assign different names
# Take care of the case of lowBound
```

```python
# Objective function
prob += 65 * x + 90 * y+ 40*z + 60*v+20*w, "Total Quality Units"
```

```python
# Constraints
prob += x*1000+y*2000+z*1500+v*2500+w*300 >=50000
prob += x*1500+y*3000+z*400+v*1000+w*100 <=30000
prob += x <=15
prob += y <=10
prob += z <=25
prob += v <=4
prob += w <=30
prob += x + y >=10
prob +=1500*x+3000*y <=18000
```

```python
# Solve the problem
prob.solve()
```

```python
# Print the results
print("We need {} units of DTV".format(x.varValue))
print("We Need {} units of ETV".format(y.varValue))
print("We Need {} units of  DN".format(z.varValue))
print("We Need {} units of  SN".format(v.varValue))
print("We Need {} units of  R".format(w.varValue))
print("Total Units are :{}".format(pulp.value(prob.objective)))
```
