```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import itertools
```

```python
df=pd.read_csv("sample_dataset_v3.csv")
df1=df.head()
df1
```

```python
df1.columns
```

The dataset contains 5 dates which are not in order. I would like to complete the dataset by filling in all the intervening dates brand and sites. I would put a zero with all such columns filled. I would have to take care that I do not introduce any row where for a particular date, brand and site there is already a row

```python
df1['Billing Date']=pd.to_datetime(df1['Billing Date'])
```

### Get the start_date and the end_date

```python
# Calculate the start date and end date
start_date=df1['Billing Date'].min().replace(day=1)
end_date=df1['Billing Date'].max()
start_date,end_date
```

### Change the start_date to 1st of that month
In this case it starts from 1st of the month. But suppose I want to start the date from 1st of the month, in case this date is not first, say 2023-01-08

```python
start_date = start_date.replace(day=1)
start_date
```

### Calculating the Date Range

```python
date_range=pd.date_range(start=start_date,end=end_date)
date_range
```

### Get Unique Brands


```python
unique_brands=df1['Brand'].unique()
unique_brands
```

### Get Unique Site Names

```python
unique_site=df1['Site Name'].unique()
unique_site
```

### Generate unique combinations of date_range, unique_brands and unique_site

```python
combinations= list(itertools.product(date_range,unique_brands, unique_site))
combinations
```

### Create a New dataframe with combinations

```python
df_combinations = pd.DataFrame(combinations,columns=['Billing Date','Brand','Site Name'])
df_combinations
```

### Keep only rows in df_combinations that don't have a match in df1

```python
anti_join = df_combinations.merge(df1[['Billing Date','Brand','Site Name']],
                                 on=['Billing Date','Brand','Site Name'],
                                 how='left',
                                 indicator=True)
anti_join
```

### Taking out those combinations Which are in the left Dataframe and not common 

```python
df_combinations=anti_join[anti_join['_merge']=='left_only']
df_combinations
```

### Dropping the _merge columns

```python
df_combinations = df_combinations.drop(columns=['_merge'])
df_combinations
```

### Getting the columns other then the combination

```python
additional_columns=df1.columns.difference(['Billing Date','Brand','Site Name'])
additional_columns
```

### Set the Sales and Other Column values to 0 for the combinations


```python
df_combinations[additional_columns]=0
df_combinations
```

### Append the combinatios dfs to the original df

```python
df1 = pd.concat([df1,df_combinations],ignore_index=True)
df1
```
