In pandas, the **pd.to_datetime()** function is used to convert a given date or time-like object to a pandas datetime object. This function can also handle various date formats and can interpret ambiguous dates. The **dayfirst** and **yearfirst** options are used to indicate the order in which the day, month, and year are specified in the input data.

**dayfirst option:**
By default, dayfirst is set to **False**. This means that the function expects the date strings in the format "yyyy-mm-dd" or "yyyy/mm/dd", where the year comes first, followed by the month and then the day.
If your date strings have the day specified first before the month, you can set dayfirst=True. This way, the function will interpret the day before the month during parsing.

**yearfirst option:**
By default, yearfirst is set to **False**. This means that the function expects the date strings in the format "dd-mm-yyyy" or "dd/mm/yyyy," where the day comes first, followed by the month, and then the year.
If your date strings have the year specified first before the day and month, you can set **yearfirst=True**. This way, the function will interpret the year before the day and month during parsing.

Here's the syntax of the pd.to_datetime() function with the dayfirst and yearfirst options:
**pd.to_datetime(arg, format=None, dayfirst=False, yearfirst=False)**

**arg:** The date-like object or list/array of date-like objects to be converted to datetime.
**format: Optional**. A string representing the date format. If not specified, pandas will try to infer the format automatically.
**dayfirst: Optional**. If set to True, pandas will interpret the day before the month during parsing.
**yearfirst: Optional**. If set to True, pandas will interpret the year before the day and month during parsing.

```python
import pandas as pd

# Date string with day first
date_string1 = "31-08-2023"
date1 = pd.to_datetime(date_string1, dayfirst=True)
print(date1)  

# Date string with year first
date_string2 = "2023-08-31"
date2 = pd.to_datetime(date_string2, yearfirst=True)
print(date2)
```

The **strftime** (string format time) method in pandas is used to convert datetime objects to custom string representations based on a specified format. This method allows you to format the date and time components of a pandas datetime object as per your requirement.

Here's an example of how to use the strftime method in pandas:

```python
import pandas as pd

# Sample data
dates = pd.Series([
    pd.to_datetime('2023-08-31 14:30:00'),
    pd.to_datetime('2023-09-15 09:45:00'),
    pd.to_datetime('2023-10-20 18:15:00')
])

# Using strftime to format the datetime objects
formatted_dates = dates.dt.strftime('%Y-%m-%d %H:%M:%S')
print(formatted_dates)
```

Here are some of the common format codes used in the strftime method to format datetime objects in Python:

%Y: 4-digit year

%y: 2-digit year (00 to 99)

%m: 2-digit month (01 to 12)

%d: 2-digit day of the month (01 to 31)

%H: 2-digit hour (00 to 23)

%I: 2-digit hour (01 to 12) (used with %p for 12-hour clock)

%M: 2-digit minute (00 to 59)

%S: 2-digit second (00 to 59)

%f: 6-digit microsecond (000000 to 999999)

%a: Abbreviated weekday name (e.g., Mon, Tue, Wed)

%A: Full weekday name (e.g., Monday, Tuesday, Wednesday)

%b: Abbreviated month name (e.g., Jan, Feb, Mar)

%B: Full month name (e.g., January, February, March)

%c: Locale's appropriate date and time representation

%p: AM or PM (used with %I for 12-hour clock)

%j: Day of the year as a 3-digit number (001 to 366)

%U: Week number of the year (Sunday as the first day of the week) as a zero-padded decimal number (00 to 53)

%W: Week number of the year (Monday as the first day of the week) as a zero-padded decimal number (00 to 53)

%w: Weekday as a decimal number (0 to 6, where 0 is Sunday)

%x: Locale's appropriate date representation

%X: Locale's appropriate time representation

%%: A literal '%' character

You can use any combination of these format codes along with other characters (such as hyphens, slashes, colons, etc.) to 
create custom date and time formats.

```python
# Sample datetime object
date = pd.to_datetime('2023-08-31 14:30:15.123456')

# %Y: 4-digit year
print(date.strftime('%Y'))   # Output: 2023

# %y: 2-digit year (00 to 99)
print(date.strftime('%y'))   # Output: 23

# %m: 2-digit month (01 to 12)
print(date.strftime('%m'))   # Output: 08

# %d: 2-digit day of the month (01 to 31)
print(date.strftime('%d'))   # Output: 31

# %H: 2-digit hour (00 to 23)
print(date.strftime('%H'))   # Output: 14

# %I: 2-digit hour (01 to 12) (used with %p for 12-hour clock)
print(date.strftime('%I'))   # Output: 02

# %M: 2-digit minute (00 to 59)
print(date.strftime('%M'))   # Output: 30

# %S: 2-digit second (00 to 59)
print(date.strftime('%S'))   # Output: 15

# %f: 6-digit microsecond (000000 to 999999)
print(date.strftime('%f'))   # Output: 123456

# %a: Abbreviated weekday name (e.g., Mon, Tue, Wed)
print(date.strftime('%a'))   # Output: Wed

# %A: Full weekday name (e.g., Monday, Tuesday, Wednesday)
print(date.strftime('%A'))   # Output: Wednesday

# %b: Abbreviated month name (e.g., Jan, Feb, Mar)
print(date.strftime('%b'))   # Output: Aug

# %B: Full month name (e.g., January, February, March)
print(date.strftime('%B'))   # Output: August

# %c: Locale's appropriate date and time representation
print(date.strftime('%c'))   # Output: Wed Aug 31 14:30:15 2023

# %p: AM or PM (used with %I for 12-hour clock)
print(date.strftime('%p'))   # Output: PM

# %j: Day of the year as a 3-digit number (001 to 366)
print(date.strftime('%j'))   # Output: 243

# %U: Week number of the year (Sunday as the first day of the week) as a zero-padded decimal number (00 to 53)
print(date.strftime('%U'))   # Output: 35

# %W: Week number of the year (Monday as the first day of the week) as a zero-padded decimal number (00 to 53)
print(date.strftime('%W'))   # Output: 35

# %w: Weekday as a decimal number (0 to 6, where 0 is Sunday)
print(date.strftime('%w'))   # Output: 3

# %x: Locale's appropriate date representation
print(date.strftime('%x'))   # Output: 08/31/23

# %X: Locale's appropriate time representation
print(date.strftime('%X'))   # Output: 14:30:15

# %%: A literal '%' character
print(date.strftime('%%'))   # Output: %
```

## Frequencies

```python
# Create a date range for 10 days starting from a specific date
date_range_daily = pd.date_range(start='2023-08-01', periods=10, freq='D')
print('Daily:',date_range_daily)

# Create a date range for 5 business days starting from a specific date
date_range_business_day = pd.date_range(start='2023-08-01', periods=5, freq='B')
print('business day:',date_range_business_day)


# Create a date range for 4 weeks starting from a specific date
date_range_weekly = pd.date_range(start='2023-08-01', periods=4, freq='W')
print('weekly:',date_range_weekly)

# Create a date range for 3 months ending on the last day of each month
date_range_month_end = pd.date_range(start='2023-08-01', periods=3, freq='M')
print('month_end:',date_range_month_end)

# Create a date range for 2 business months ending on the last business day of each month
date_range_business_month_end = pd.date_range(start='2023-08-01', periods=2, freq='BM')
print('business_month:',date_range_business_month_end)

# Create a date range for 2 quarters ending on the last day of each quarter
date_range_quarterly = pd.date_range(start='2023-01-01', periods=2, freq='Q')
print('quarterly"',date_range_quarterly)

# Create a date range for 5 years ending on the last day of each year
date_range_year_end = pd.date_range(start='2023-01-01', periods=5, freq='A')
print('year end',date_range_year_end)
```

## Offsets
Offsets are **variations on a base frequency**. For example we can take the business days range, and add an offset such that instead of every business day,data are included for every other business day.

We do this by putting a multiplying value before the base frequency

```python
# every other business day during the week of Jan1,2017
print(pd.date_range('2017-01-01','2017-01-07',freq='2B'))

# First Thursday of each month in the year 2017
print((pd.date_range('2017-01-01','2017-12-31',freq='WOM-1THU')))

# Third Friday of Every Month
print(pd.date_range('2017-01-01','2017-12-31',freq = 'WOM-3FRI'))
```

In this example, we want all our dates to start from a common 0 day. There are multiple steps to this process.

1. Since not every date is listed, we need to create a date range of all the dates in our dataset
2. We need to calculate the difference between the earliest date in our data set, and the earliest valid ( non NAN) date in each column
3. We can then shift each of the columns by this calculated value

Let's begin with a fresh copy of the Ebola dataset

```python
ebola = pd.read_csv("country_timeseries.csv",index_col='Date',parse_dates=['Date'])
ebola.head().iloc[:,:4]
```

```python
ebola.tail().iloc[:,:4]
```

```python
# create the date_range
new_idx = pd.date_range(ebola.index.min(),ebola.index.max())
new_idx
```

```python
# The dates are not in the order that we want we have to reverse it
new_idx = reversed(new_idx)
new_idx
```

```python
ebola = ebola.reindex(new_idx)
```

```python
ebola.head()
```

To get the last valid index, we use the Series method called **last_valid_index**, which returns the index of the last non-missing or non-null value. An analogous method called **first_valid_index** returns the first non-missing or non-null values. Since we want to perform this calculations across all columns, we can use the apply method

```python
last_valid = ebola.apply(pd.Series.last_valid_index)
last_valid
```

```python
# Now we get the earliest date in our data set
earliest_date = ebola.index.min()
# we then subtract this date from the last valid date
shift_values = last_valid -earliest_date
shift_values
```

```python
ebola_dict = {}
for idx, col in enumerate(ebola):
    d = shift_values[idx].days
    shifted = ebola[col].shift(d)
    ebola_dict[col] = shifted
```

**ebola_dict = {}**
This line initializes an empty Python dictionary named ebola_dict

**for idx, col in enumerate(ebola):**
This line starts a for loop that iterates over the ebola object. As mentioned earlier, ebola seems to be some iterable (e.g., a list or a pandas DataFrame). The enumerate() function is used to loop through the elements of ebola while keeping track of the index and the value of each element.

idx: This variable will hold the index of the current element being processed.
col: This variable will hold the value of the current element being processed.
During each iteration of the loop, idx will be the index of the element in the ebola object, and col will be the value of the element.

**d = shift_values[idx].days**
This line retrieves the number of days from the shift_values list (or another similar object) using the current index idx and assigns it to the variable d. shift_values appears to be a list of timedelta objects or something similar.

**shifted = ebola[col].shift(d)**
This line performs a shift operation on the col column of the ebola DataFrame. The shift() method is a pandas DataFrame method that shifts the data in the specified column by the number of periods specified by d. It moves the data values d positions up (for positive d) or down (for negative d) in the column. If d is positive, it brings older dates into the current row, and if d is negative, it brings future dates into the current row.

The result of this operation is stored in the shifted variable, which becomes a new pandas Series with the shifted values.

**ebola_dict[col] = shifted**
This line adds the shifted Series to the ebola_dict dictionary, with the column name col as the key. This is done so that the shifted Series for each column in the ebola DataFrame is stored in the ebola_dict dictionary with the corresponding column name.

The loop continues to iterate over all the columns in the ebola DataFrame, performing the shift operation on each column and storing the shifted Series in the ebola_dict dictionary.

Overall, this code takes a DataFrame ebola with some time-based data and applies a shift operation to each column based on the number of days specified in the shift_values list (or another similar object). The result is stored in a dictionary ebola_dict, where each key corresponds to the column name, and each value is the shifted Series for that column.


```python
ebola_shift = pd.DataFrame(ebola_dict)
# The dict objects are unordered, but we can pass in the original ebola ccolumns to reorder the values back
ebola_shift = ebola_shift[ebola.columns]
# The last row in each column now has a value, that is, the columns have been shifted down appropriatedly
ebola_shift.tail()
```

Finally since the indices are no longer valid across each row, we can remove them, and then assign the correct index, which is the Day

Note that Day no longer represents the first dayof the entire outbreak, but rather the first day of an outbreak for the given country

```python
ebola_shift.index = ebola_shift['Day']
ebola_shift = ebola_shift.drop(['Day'],axis=1)
ebola_shift.tail()
```

## Resampling
Resampling convers a datetime from one frequency to another frequency. Three types of resampling can occur:

1. **Downsampling**: from a higher frequency to a lower frequency (eg. daily to monthly)
2. **Upsampling**: from a lower frequency to a higher frequency ( eg. monthly to daily)
3. **No Change**: frequency does not change ( eg. every first Thursday of the month to Last Friday of the month)

```python
# downsample daily values to monthly values
# since we have multiple values, we need to segregate the results
# here we will use the mean
down = ebola.resample('M').mean()
down.iloc[:5,:5]
```

```python
# here we will upsample our downsampled value
# notice how missing dates are populated
# but they are filled in missing values
up = down.resample('D').mean()
print(up.iloc[:5,:5])
```
