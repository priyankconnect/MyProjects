```python
# Create an empty graph with no nodes and no edges
```

```python
import networkx as nx
G = nx.Graph()
```

```python
# By definition, a Graph is a collection of nodes (vertices) along with identified pairs 
#of nodes (called edges, links, etc)
```

```python
# you can add nodes one at a time
G.add_node(1)
```

```python
# You can also add nodes from a list
G.add_nodes_from([2, 3])
```

```python
# You can add nodes with node attributes
G.add_nodes_from([
    (4, {"color": "red"}),
    (5, {"color": "green"}),
])
```

```python
# You can incorporate the nodes from another graph
H = nx.path_graph(10) # It creates a graph with nodes in straight line
G.add_nodes_from(H)
```

```python
# You could use H as a node as well
G.add_node(H)
```

```python
# Edges
# G can also be grown by adding one edge at a time,
G.add_edge(1, 2)
e = (2, 3)
G.add_edge(*e)  # unpack edge tuple*
```

```python
# By adding a list of edges
G.add_edges_from([(1, 2), (1, 3)])
```

```python
# You can add edges from the edge of other graphs
G.add_edges_from(H.edges)
```

```python
# Removing all edges and nodes
G.clear()
```

```python
G.add_edges_from([(1, 2), (1, 3)])
G.add_node(1)
G.add_edge(1, 2)
G.add_node("spam")        # adds node "spam"
G.add_nodes_from("spam")  # adds 4 nodes: 's', 'p', 'a', 'm'
G.add_edge(3, 'm')
```

```python
# How to find the number of nodes
G.number_of_nodes()
```

```python
# How to find the number of edges
G.number_of_edges()
```

```python
# there are four basic graph properties: G.nodes, G.edges, G.adj and G.degree
#These are set-like views of the nodes, edges, neighbors (adjacencies), and degrees of nodes in a graph.
# they are dict like so you need to convert them to list
```

```python
list(G.nodes)
```

```python
list(G.edges)
```

```python
list(G.adj[1])
```

```python
list(G.neighbors(1)) # Same as one before
```

```python
G.degree[1]  # the number of edges incident to 1
```

```python
# One can see the edges of an edge
G.edges([2, 'm'])
```

```python
# One can see the degree of an edge
G.degree([2, 3])
```

```python
# remove elements from a graph
G.remove_node(2)
list(G.nodes)
```

```python
G.remove_nodes_from("spam")
list(G.nodes)
```

```python
G.remove_edge(1, 3)
list(G.nodes)
```

```python
list(G.edges)
```

```python
# Create a graph from the edge list
edgelist = [(0, 1), (1, 2), (2, 3)]
H = nx.Graph(edgelist)
list(H.edges())
```

```python
# create a Graph from adjacency dictionary
adjacency_dict = {0: (1, 2), 1: (0, 2), 2: (0, 1)}
H = nx.Graph(adjacency_dict)
list(H.edges())
```

## Accessing Edges and Neighbors

```python
# access to edges and neighbors is possible using subscript notation.
G = nx.Graph([(1, 2, {"color": "yellow"})])
G[1]  # same as G.adj[1]
```

```python
G[1][2]
```

```python
G.edges[1, 2]
```

```python
# You can get/set the attributes of an edge using subscript notation if the edge already exists.
G.add_edge(1, 3)
G[1][3]['color'] = "blue"
G.edges()
```

```python
G.edges[1, 2]['color'] = "red"
G.edges[1, 2]
```

```python
# Attributes such as weights, labels, colors, or whatever Python object you like, 
# can be attached to graphs, nodes, or edges.
#It's important to note that the attribute 'day'='Friday' doesn't affect the structure 
#or the graphical representation of the graph. It's simply a piece of data associated with the graph. 
G = nx.Graph(day="Friday")
G.graph
```

```python
# You can modify it later
G.graph['day'] = "Monday"
G.graph
```

```python
# Node Attributes
G.add_node(1, time='5pm')
G.add_nodes_from([3], time='2pm')
G.nodes[1]
```

```python
G.nodes[1]['room'] = 714
```

```python
G.nodes.data()
```

```python
# Edge Attributes
G.add_edge(1, 2, weight=4.7 )
G.add_edges_from([(3, 4), (4, 5)], color='red')
G.add_edges_from([(1, 2, {'color': 'blue'}), (2, 3, {'weight': 8})])
G[1][2]['weight'] = 4.7
G.edges[3, 4]['weight'] = 4.2
```
