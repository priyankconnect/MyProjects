### Principal Component Analysis

PCA is a data reduction technique that transforms a larger number of correlated variables into a much smaller set of correlated variables such that the maximum information ( variance) is retained is called Principal Components. 

The first component has the highest variance followed by second, third and so on. The components are uncorrelated. 

It is always performed on a symmetric correlation or covariance matrix. This means the matrix should be numeric and have standardized data. 

In general for a n x p dimensional data, min(n-1,p) PC can be constructed

The function named PCA() available in sklearn.decomposition is used to perform PCA by specifying the number of components

```python
# Principal Component Analysis
# Importing necessary libraries and loading data
from sklearn.decomposition import PCA
from sklearn.datasets import make_friedman1
x,y=make_friedman1()
```

This example considers Friedman dataset available in sklearn.datasets library. The make_friedman1() function returns two datasets: for independent variables and dependent variables. 

Link: https://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_friedman1.html

The dataset corresponding to independent variables is stores in x , whereas the dataset corresponding to dependent variable is stored in y. 

Since we do not consider the response variable in unsupervised learning, we will consider only the "x" dataset. 

The dataset corresponding to independent variables have 100 observations and 10 variables.

```python
# Performing principle component analysis
pca=PCA(n_components=4)
pca.fit(x)
x_transformed=pca.transform(x)
print("Original shape of Friedman data set:",x.shape)
print("Reduced shape of Friedman data set:",x_transformed.shape)
print("Shape of Principal components:",pca.components_.shape)
print("Details of Principal Components:\n",pca.components_.round(3))
print("Component_wise explained variance:\n",pca.explained_variance_ratio_.round(3))
```

Interpretation:

The choice of how many principal components to use depends on the problem at hand and how much variance you're willing to retain versus how much dimensionality you want to reduce. In this case, using 4 components captures a little over half the total variance.
To decide whether this is sufficient, you would typically look at a scree plot or cumulative explained variance plot to see how adding more components affects the total explained variance.
The actual principal component vectors (the rows of your matrix) give you an understanding of how the original features combine to form new, orthogonal axes in the data. Features with larger absolute values in these vectors are more influential in defining that component.

In summary, PCA has allowed you to reduce your data from 10 features to 4 principal components, while retaining about 52.5% of the information in the dataset. You can use these components in further analyses, visualizations, or machine learning models.
