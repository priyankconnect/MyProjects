## Stock Price Simulations in Code

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Number of Time steps
T=1000

# Initial Price

P0=10

# Drift

mu=0.001
```

```python
# Last log price

last_p = np.log(P0) # this variable will be updated since the current price will always be dependent on last price

log_returns= np.zeros(T) # creating array to enter log return
prices=np.zeros(T)
for t in range(T):  # creating array to enter log prices 
    #sample a log return
    
    r = 0.01*np.random.randn() # a sample with zero mean and standard deviation 0.01
    
    #compute the new log price
    p = last_p+mu+r # equal to the old log price plus the mu plus the random noise. 
    
    # store and return the price
    log_returns[t] = r + mu
    prices[t]=np.exp(p) # converting again from log price to price
    
    # assign last_p
    last_p = p  # we assign the current value of p to the last p for the next iteration
```

```python
plt.figure(figsize=(20,8))
plt.plot(prices)
plt.title('Simulated Prices');
```
