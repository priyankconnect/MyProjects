```python
from statsmodels.graphics.tsaplots import plot_pacf,plot_acf
```

```python
import numpy  as np
import matplotlib.pyplot as plt
```

```python
x0=np.random.randn(1000)
```

```python
plt.plot(x0)
```

```python
fig,ax=plt.subplots(figsize=(10,5))
plot_pacf(x0,ax=ax);
```

```python
# Creating an AR1 process
x1=[0]
for i in range(1000):
    x=0.5*x1[-1]+0.1*np.random.randn()
    x1.append(x)
    
x1=np.array(x1)
```

```python
plt.plot(x1);
```

```python
fig,ax=plt.subplots(figsize=(10,5))
plot_pacf(x1,ax=ax);
```

```python
# Creating an AR1 process
x1=[0]
for i in range(1000):
    x=-0.5*x1[-1]+0.1*np.random.randn()
    x1.append(x)
    
x1=np.array(x1)
```

```python
plt.plot(x1);
```

```python
fig,ax=plt.subplots(figsize=(10,5))
plot_pacf(x1,ax=ax);
```

```python
# Generating an AR2 process
x2=[0,0]
for i in range(1000):
    x=0.5*x2[-1]-0.3*x2[-2]+0.1*np.random.randn()
    x2.append(x)
x2=np.array(x2)
```

```python
plt.plot(x2);
```

```python
fig,ax=plt.subplots(figsize=(10,5))
plot_pacf(x2,ax=ax);
```

```python
# Generate an AR5 process
x5=[0,0,0,0,0]
for i in range(1000):
    x=0.5*x5[-1]-0.3*x5[-2]-0.6*x5[-5]+0.1*np.random.randn()
    x5.append(x)
x5=np.array(x5)
```

```python
plt.plot(x5);
```

```python
fig,ax=plt.subplots(figsize=(10,5))
plot_pacf(x5,ax=ax);
```

### Plotting ACF

```python
# iid Noise
fig,ax=plt.subplots(figsize=(10,5))
plot_acf(np.random.randn(1000),ax=ax);
```

```python
# Generating an MA1 process
errors=0.1*np.random.randn(1000)
ma1=[]
for i in range(1000):
    if i>=1:
        x=0.5*errors[i-1]+errors[i]
    else:
        x=errors[i]
    ma1.append(x)
ma1=np.array(ma1)
```

```python
plt.plot(ma1);
```

```python
fig,ax=plt.subplots(figsize=(10,5))
plot_acf(ma1,ax=ax);
```

```python
# Generating an MA2 process
errors=0.1*np.random.randn(1000)
ma2=[]
for i in range(1000):
    x=0.5*errors[i-1]-0.3*errors[i-2]+errors[i]
    ma2.append(x)
ma2=np.array(ma2)
```

```python
plt.plot(ma2);
```

```python
fig,ax=plt.subplots(figsize=(10,5))
plot_acf(ma2,ax=ax);
```

```python
# Generating an MA3 process
errors=0.1*np.random.randn(1000)
ma3=[]
for i in range(1000):
    x=0.5*errors[i-1]-0.3*errors[i-2]+0.7*errors[i-3]+errors[i]
    ma3.append(x)
ma3=np.array(ma3)
```

```python
plt.plot(ma3)
```

```python
fig,ax=plt.subplots(figsize=(10,5))
plot_acf(ma3,ax=ax);
```

```python
# Generating an MA6 process
ma6=[]
for i in range(1000):
    x=0.5*errors[i-1]-0.3*errors[i-2]+0.7*errors[i-3]+\
    0.2*errors[i-4]-0.8*errors[i-5]-0.9*errors[i-6]+errors[i]
    ma6.append(x)
ma6=np.array(ma6)
```

```python
plt.plot(ma6)
```

```python
fig,ax=plt.subplots(figsize=(10,5))
plot_acf(ma6,ax=ax);
```
