```python
import xlwings  as xw
import warnings
import pandas as pd
import numpy as np
warnings.filterwarnings('ignore')

store = pd.read_excel("storestock.xlsx",sheet_name = 'replenishment') # gap file
dc = pd.read_excel("dc_stock.xlsx",sheet_name ='soh')
stsoh1 = pd.read_excel("stsoh.xlsx",sheet_name ='soh')

dc1 = dc[dc['qty']>0]
final = pd.DataFrame()
print("Stock Before Allocation:",dc1.qty.sum())
print("========================")

for a in store['STORE'].unique():
    print(a)
    dc1 = dc1[dc1['qty']>0]
    dcO = dc1.sample(frac=1)
    print("Stock for allocation for the store:",dcO.qty.sum())

    storeA = store[store['STORE']==a]
    store1 = storeA[storeA['gap']>0]
    store1['gap'] = store1['gap'].apply(np.ceil)
    store2 = store1.copy()
    print('gap:',store2.gap.sum())
    store2.shape

    store2.head()
    store2.shape

    if store2.empty:
        print('Exact Match: No allocation Needed')
        selected_df = pd.DataFrame()
        selected_dfplus = pd.DataFrame()
        selected_dfminus = pd.DataFrame()
        flag3 = False
    else:
        flag3= True

    if flag3:
        all_nan_rows = store2[store2.isna().any(axis=1)]

    if flag3:
        merged_df = store2.merge(dcO,on = ['Brand','Craft','Price_Level'],how = 'inner')


    if flag3:
        stsoh = stsoh1[stsoh1['Store']==a]

    if flag3:
        def select_top_n(df_grouped,stsoh):
          total_alloc = df_grouped.iloc[0]["gap"]
          df_grouped = df_grouped[~df_grouped['Article'].isin(stsoh['Article'])]
          selected_rows = df_grouped.iloc[:int(total_alloc)]
          return selected_rows
    if flag3:
        grouped = merged_df.groupby(["Brand","Craft","Price_Level"])

    if flag3:
        selected_df = grouped.apply(select_top_n,stsoh)
    if flag3:
        merged_df4_1 = pd.merge(selected_df,dcO,on='Article',how='inner')

    if flag3:
        selected_df = selected_df.reset_index(drop=True)
        print("Allocation at Target Price_Level",selected_df.Article.count())

    if flag3:
        merged_df4_1['qty_y'] = merged_df4_1['qty_y']-1
    if flag3:
        dcO.loc[dcO['Article'].isin(selected_df['Article']),'qty'] = 0

    if flag3:
        selected1 = selected_df.groupby(['Brand','Craft','Price_Level'])['Article'].count().reset_index()

    if flag3:
        merged_test = store2.merge(selected1,on=['Brand','Craft','Price_Level'],how ='inner')
        merged_test['diff'] = merged_test['gap']-merged_test['Article']
    if flag3:
        store2 = merged_test[(merged_test['diff'].isnull()) | (merged_test['diff']>0)]
    if flag3:
        if store2.empty:
            print('Price Level +1 or +2 Allocation: No Allocation Needed')
            store2_2 = pd.DataFrame()
            selected_dfplus = pd.DataFrame()
            selected_dfminus = pd.DataFrame()
            flag2 = False
        else:
            flag2 = True
    if flag3:
        if flag2:
            store2_2 = store2.copy()
            store2_2.loc[(store2_2['diff']>0) & (store2_2['diff'].notna()),'gap'] = store2_2['diff']
    if flag3:
        if flag2:
            store2_2 = store2_2.drop(['Article','diff'],axis=1)
    if flag3:
        if flag2:
            store2_2['Price_Level'] = store2_2['Price_Level']+1
    if flag3:
        if flag2:
            merged_df = store2_2.merge(dcO,on=['Brand','Craft','Price_Level'],how ='inner')
    if flag3:
        if flag2:
            stsoh = stsoh1[stsoh1['Store']==a]
    if flag3:
        if flag2:
            def select_top_n(df_grouped,stsoh):
                total_alloc = df_grouped.iloc[0]["gap"]
                df_grouped = df_grouped[~df_grouped['Article'].isin(stsoh['Article'])]
                selected_rows = df_grouped.iloc[:int(total_alloc)]
                return selected_rows
    if flag3:
        if flag2:
            grouped = merged_df.groupby(["Brand","Craft","Price_Level"])
            
    if flag3:
        if flag2:
            selected_dfplus = grouped.apply(select_top_n,stsoh)
    if flag3:
        if flag2:
            selected_dfplus = selected_dfplus.reset_index(drop=True)
            print("Allocation at Price_Level+1",selected_dfplus.Article.count())
    if flag3:
        if flag2:
            merged_df4_1 = pd.merge(selected_dfplus,dcO,on='Article',how='left')
            merged_df4_1.head()
    if flag3:
        if flag2:
            merged_df4_1['qty_y']-1
            merged_df4_1.head()
    if flag3:
        if flag2:
            dcO = dcO[dcO['qty']>0]
    if flag3:
        if flag2:
            selected1 = selected_dfplus.groupby(['Brand','Craft','Price_Level'])['Article'].count().reset_index()
    if flag3:
        if flag2:
            merged_test = store2_2.merge(selected1,on=['Brand','Craft','Price_Level'],how='inner')
            merged_test['diff']=merged_test['gap']-merged_test['Article']
    if flag3:
        if flag2:
            store2_2 = merged_test[(merged_test['diff'].isnull()) | (merged_test['diff']>0)]
    if flag3:
        if flag2:
            if store2_2.empty:
                print('Price Level -1 Allocation: No Allocation Could be done')
                store2_3 = pd.DataFrame()
                selected_dfminus = pd.DataFrame()
                flag = False
            else:
                flag = True
    if flag3:
        if flag2:
            store2_3 = store2_2.copy()
            store2_3.loc[(store2_3['diff']>0) & (store2_3['diff'].notna()),'gap'] = store2_3['diff']
    if flag3:
        if flag2:
            store2_3 = store2_3.drop(['Article','diff'],axis=1)
    if flag3:
        if flag2:
            store2_3['Price_Level']=store2_3['Price_Level']-2


    if flag3:
        if flag2:
            merged_df = store2_3.merge(dcO, on =['Brand','Craft','Price_Level'],how='inner')
    if flag3:
        if flag2:
            stsoh = stsoh1[stsoh1['Store']==a]
    if flag3:
        if flag2:
            def select_top_n(df_grouped,stsoh):
                total_alloc = df_grouped.iloc[0]["gap"]
                df_grouped = df_grouped[~df_grouped['Article'].isin(stsoh['Article'])]
                selected_rows = df_grouped.iloc[:int(total_alloc)]
                return selected_rows

    if flag3:
        if flag2:
            grouped = merged_df.groupby(['Brand','Craft','Price_Level'])
            
    if flag3:
        if flag2:
            selected_dfminus = grouped.apply(select_top_n,stsoh)
    if flag3:
        if flag2:
            selected_dfminus = selected_dfminus.reset_index(drop=True)
            print("Allocation at Price_Level-1:",selected_dfminus.Article.count())
    if flag3:
        if flag2:
            merged_df4_1['qty_y']= merged_df4_1['qty_y']-1
            merged_df4_1.head()
    if flag3:
        if flag2:
            dcO.loc[dcO['Article'].isin(selected_dfminus['Article']),'qty']=0
    if flag3:
        if flag2:
            dcO = dcO[dcO['qty']>0]
    if flag3:
        if flag2:
            selected1 = selected_dfminus.groupby(['Brand','Craft','Price_Level'])['Article'].count().reset_index()
    if flag3:
        if flag2:
            merged_test = store2_3.merge(selected1,on=['Brand','Craft','Price_Level'],how = 'inner')
            merged_test['diff'] = merged_test['gap']- merged_test['Article']
    if flag3:
        if flag2:
            store2_3 = merged_test[(merged_test['diff'].isnull()) | (merged_test['diff']>0)]

    sel = pd.concat([selected_df,selected_dfplus,selected_dfminus])
    sel = sel.reset_index(drop = True)

    short_replenishment = store1.gap.sum()-sel.shape[0]

    if not sel.empty:
        output= sel[['STORE','Brand','Craft','Price_Level','Article']]
        output['qty']=1
    else:
        output = pd.DataFrame()
    if not output.empty:
        dc1.loc[dc1['Article'].isin(output['Article']),'qty'] -=1
    else:
        print("No Final Allocation for this store")

    print("Total allocated:",output.qty.sum())
    final = pd.concat([final,output])
    print("Stock after allocation:",dc1.qty.sum())
    print("======================================")
print("================================")
print("================================")
print("================================")
print("Final Qty Allocated:",final.qty.sum())
newfile = "replenishmentoutput.xlsx"
newstock = "dcstocknew.xlsx"

with pd.ExcelWriter(newfile)as writer:
    final.to_excel(writer,sheet_name='replenishment',index=False)
with pd.ExcelWriter(newstock) as writer:
    dc1.to_excel(writer,sheet_name='dcsoh',index=False)
```
