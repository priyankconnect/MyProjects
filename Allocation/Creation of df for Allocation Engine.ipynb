```python
import pandas as pd
import numpy as np
import random
import string

# Generate unique style codes
def generate_style_code():
    prefix = ''.join(random.choices(string.ascii_uppercase, k=9))
    suffix = ''.join(random.choices(string.digits, k=5))
    return prefix + suffix

unique_style_codes = set()
while len(unique_style_codes) < 4560:
    unique_style_codes.add(generate_style_code())

# Repeat the unique style codes to cover all articles
all_style_codes = np.tile(list(unique_style_codes), 10330 // 4560 + 1)[:10330]

# Shuffle the style codes to ensure randomness
np.random.shuffle(all_style_codes)

# Create a DataFrame
df = pd.DataFrame({
    'Article': ['A' + str(i) for i in range(1, 10331)],
    'Style_Code': all_style_codes
})

import pandas as pd
import numpy as np
import random
import string

# Sample data generation (from the previous example)
def generate_style_code():
    prefix = ''.join(random.choices(string.ascii_uppercase, k=9))
    suffix = ''.join(random.choices(string.digits, k=5))
    return prefix + suffix

unique_style_codes = set()
while len(unique_style_codes) < 4560:
    unique_style_codes.add(generate_style_code())

all_style_codes = np.tile(list(unique_style_codes), 10330 // 4560 + 1)[:10330]
np.random.shuffle(all_style_codes)

df = pd.DataFrame({
    'Article': ['A' + str(i) for i in range(1, 10331)],
    'Style_Code': all_style_codes
})

# Create Article1 column
# Create a prefix for each unique Style_Code
unique_prefixes = ["22" + str(i).zfill(7) for i in range(1, len(unique_style_codes) + 1)]
style_to_prefix = dict(zip(unique_style_codes, unique_prefixes))

# Function to generate Article1 based on Style_Code occurrence
style_code_counter = {}
def generate_article_code(style_code):
    if style_code not in style_code_counter:
        style_code_counter[style_code] = 1
    else:
        style_code_counter[style_code] += 1
    return style_to_prefix[style_code] + str(style_code_counter[style_code]).zfill(3)

df['Article1'] = df['Style_Code'].apply(generate_article_code)

# Create unique crafts
crafts = ['Craft_' + str(i) for i in range(1, 120)]

# Map each unique style code to a craft
style_to_craft = {style: random.choice(crafts) for style in unique_style_codes}

# Add the mapped crafts to the DataFrame
df['Craft'] = df['Style_Code'].map(style_to_craft)

# Generate MRPs from a normal distribution
mean_mrp = (29999 + 599) / 2
std_dev_mrp = (29999 - 599) / 6  # Assuming 99.7% of data within 599 and 29999

mrps = np.random.normal(mean_mrp, std_dev_mrp, len(unique_style_codes))
mrps = np.clip(mrps, 599, 29999)  # Ensure MRPs are within the desired range
mrps = np.round(mrps).astype(int)

# Map each unique style code to an MRP
style_to_mrp = dict(zip(unique_style_codes, mrps))

# Add the mapped MRPs to the DataFrame
df['MRP1'] = df['Style_Code'].map(style_to_mrp)

# Define bin edges and labels
bin_edges = list(range(0, 31000, 1000))
bin_labels = [f"{bin_edges[i]}-{bin_edges[i+1]}" for i in range(len(bin_edges)-1)]

# Create MRP bins
df['MRP_Bin'] = pd.cut(df['MRP1'], bins=bin_edges, labels=bin_labels, right=False)

# Define the bins and labels for the price column
bins = [0, 5000, 15000, np.inf]
labels = [1, 2, 3]

df['price'] = pd.cut(df['MRP1'], bins=bins, labels=labels, right=False)

import pandas as pd
import numpy as np

# Assuming df is already defined from the previous steps

# Generate quantities from a normal distribution
quantities = np.random.normal(6.65, 10, len(df))
quantities = np.clip(quantities, 1, 251)
quantities = np.round(quantities).astype(int)

# Adjust values to try and match the provided quartiles
sorted_quantities = np.sort(quantities)
q1_index = int(0.25 * len(df))
q2_index = int(0.5 * len(df))
q3_index = int(0.75 * len(df))

sorted_quantities[q1_index:q2_index] = 2
sorted_quantities[q2_index:q3_index] = 4
sorted_quantities[q3_index:] = 6

# Shuffle the adjusted quantities to randomize the order
np.random.shuffle(sorted_quantities)

df['quantity'] = sorted_quantities


# Define the brands and their respective percentages
brands = ['Brand_A', 'Brand_B', 'Brand_C', 'Brand_D']
percentages = [0.0677, 0.0092, 0.9043, 0.0189]

# Calculate the number of style codes for each brand
brand_counts = [int(p * len(unique_style_codes)) for p in percentages]

# Ensure the total counts match the number of unique style codes
brand_counts[-1] += len(unique_style_codes) - sum(brand_counts)

# Assign each unique style code to a brand
assigned_brands = np.repeat(brands, brand_counts)
np.random.shuffle(assigned_brands)

style_to_brand = dict(zip(unique_style_codes, assigned_brands))

# Map the brand assignment to the main DataFrame
df['Brand'] = df['Style_Code'].map(style_to_brand)

# Generate vendor percentages
num_vendors = 57
avg_percentage = 0.0175
total_percentage = 1.0
vendor_percentages = []

for _ in range(num_vendors - 1):
    max_percentage = (total_percentage - sum(vendor_percentages) - (num_vendors - len(vendor_percentages) - 1) * 0.0001)
    max_percentage = min(max_percentage, 0.1197)
    min_percentage = 0.0001
    percentage = np.random.uniform(min_percentage, max_percentage)
    vendor_percentages.append(percentage)

# Ensure the total percentages sum up to 100%
vendor_percentages.append(total_percentage - sum(vendor_percentages))

# Calculate the number of style codes for each vendor
vendor_counts = [int(p * len(unique_style_codes)) for p in vendor_percentages]

# Ensure the total counts match the number of unique style codes
vendor_counts[-1] += len(unique_style_codes) - sum(vendor_counts)

# Assign each unique style code to a vendor
vendors = ['Vendor_' + str(i) for i in range(1, num_vendors + 1)]
assigned_vendors = np.repeat(vendors, vendor_counts)
np.random.shuffle(assigned_vendors)

style_to_vendor = dict(zip(unique_style_codes, assigned_vendors))

# Map the vendor assignment to the main DataFrame
df['Vendor'] = df['Style_Code'].map(style_to_vendor)



print(df.head())  # Display the first few rows
print("============")

# Filter rows with duplicated Style_Code
duplicated_rows = df[df['Style_Code'].duplicated(keep=False)].sort_values('Style_Code')

print(duplicated_rows)
```
